```python
#Using Array
class ArrayStack:
    def __init__(self):
      self.size=0
      self.arr=[None]*100

    #Returns The number of items on the stack
    def size(self):
      return self.size

    #Returns true if the stack is empty
    def isEmpty(self):
      if(self.size==0):
        return True
      return False

    #Pushes the new item on the stack
    def push(self,e):
      if len(self.arr)==self.size:
        return "Overflow"
      elif len(self.arr)>self.size:
        self.arr[self.size]=e
        self.size+=1


    
    #Pops the item on the top of the stack
    def pop(self):
      if self.size==0:
        return None
      else:
        temp = self.arr[self.size-1]
        self.arr[self.size-1]=None
        self.size-=1
        return temp
      

    #Returns the item on the top of the stack
    def peek(self):
      if self.size==0:
        return None
      else:
        temp = self.arr[self.size-1]
        return temp
      

    #Print all items of the stack 
    def stackPrint(self):
      for i in range(self.size):
          print(self.arr[i],end=" ")
```

```python

class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
#Using Linked List
class ListStack:
    def __init__(self):
      self.size=0
      self.top=None

    #Returns The number of items on the stack
    def size(self):
      return self.size

    #Returns true if the stack is empty
    def isEmpty(self):
      if(self.size==0):
        return True
      return False

    #Pushes the new item on the stack
    def push(self,e):
        n= Node(e, None)
        n.next = self.top
        self.top= n
        self.size+=1
    
    #Pops the item on the top of the stack
    def pop(self):
        if self.size>0:
          elem=self.top.element
          self.top=self.top.next
          self.size-=1
        return elem
    #Returns the item on the top of the stack
    def peek(self):
        if self.top!=None:
          elem=self.top.element
          
        return elem

    #Print all items of the stack 
    def listPrint(self):
      t=self.top
      while(t!=None):
          print(t.element,end="")
          t=t.next
```
