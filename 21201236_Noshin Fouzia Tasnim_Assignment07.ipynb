```python
class KeyIndex:
  k=None
  def __init__(self,a):
    
    min=0
    max=a[0]
    for i in a:
      if i<min:
        min=i
    for i in a:
      if i>max:
        max=i
    self.x= -1*min
    self.k=[0]*(self.x+max+1)
    for i in a:
      if self.k[i+self.x]==0:
        self.k[i+self.x]=1
      else:
        self.k[i+self.x]+=1        
    
  def search(self,val):
    if val<0 or val>len(self.k):
      return False
    elif self.k[val+self.x]>0:
      return True
    else:
      return False

  def sort(self):
    length=0
    for i in self.k:
        length+=i
    b=[0]*length
    idx=0
    for i in range(len(self.k)):
      for j in range(self.k[i]):
        b[idx]=i-self.x
        idx+=1
    return b
      
      
a=[9,5,3,-3,-2,-3,0,9]
b=KeyIndex(a) 
print(b.search(100))
print(b.search(1))  
print(b.search(3)) 
print(b.sort())
```

```python
def hash(a):
  hashtable=[None]*9
  vowel="AEIOU"
  num="0123456789"
  
  
  for i in a :
    
    sum=0
    count=0
    for x in i:
      if x in num:
        sum+=int(x)
      elif x not in vowel:
        count+=1
    
    hash_function= (count*24 + sum) %9
    idx=hash_function
    if hashtable[idx]==None:
      hashtable[idx]=i
    else:
      while hashtable[idx]!=None:
        idx=(idx+1)%len(hashtable)
      hashtable[idx]=i
  return hashtable


a=hash(["ST1E89B8A32","ST1E89B8A32"])
print(a)
b = hash(["ST1E89B8A32","ST1E89B8A311","ST1E89B8A11111","ST1E89B8A32","ST1E89B8A311","ST1E89B8A11"])
print(b)
c=hash(["ST1E89B8A32"])
print(c)



```
