```python
mul_arr = [[""]*10,[""]*10]

txt_1= input("Enter the first text:")
txt_2= input("Enter the second text:")

if len(txt_1)>10 or len(txt_2)>10:
  print("Invalid input size")

else:

  for i in range(len(txt_1)):
    mul_arr[0][i]=txt_1[i]

  for i in range(len(txt_2)):
    mul_arr[1][i]=txt_2[i]

  for i in range(len(txt_1)):
    if 65<=ord(txt_1[i])<=90:
      st_1=i
  og_1=st_1
  print(f"Starting Index: {og_1} \nStart index character: {txt_1[st_1]} ")

  for i in range(len(txt_2)):
    if 65<=ord(txt_2[i])<=90:
      st_2=i
  og_2=st_2
  print(f"Starting Index: {og_2} \nStart index character: {txt_2[st_2]} ")

  def rotate_left(source): 
    for i in range(0,len(source)-1):
      temp = source[i]
      source[i]=source[i+1]
      source[i+1]=temp
    s1= ""
    for x in source:
      s1+=x
    print(s1)
    return source
  def right_rotate(source):
    temp= source[len(source)-1]
    for i in range(len(source)-1,0,-1):
      source[i]= source[i-1]
      source[i-1]= temp
    s2= ""
    for x in source:
      s2+=x
    print(s2)
    return source
  
  while True:
    user_prompt= input("Enter any character:")
    if user_prompt=="Q" or user_prompt=="q":
      print("Your input is valid. Press any key and then press enter to continue!!!")
      print(f"Top Board Start Character: {txt_1[og_1]} ")
      print(f"Top Board Start Index: {og_1} ")
      print(f"Bottom Board Start Character: {txt_2[og_2]} ")
      print(f"Bottom Board Start Index: {og_2} ")
      break
    else:
      mul_arr[0]=rotate_left(mul_arr[0])
      mul_arr[1]=right_rotate(mul_arr[1])
  
    
```
