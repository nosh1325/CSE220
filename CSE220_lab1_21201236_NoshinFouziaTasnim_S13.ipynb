```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):  # Hint, You can write a function for left shift once and then use it
  for i in range(len(source)-k):
    source[i] = source[i+k]
    source[i+k]=0

  return source



# Test 02: Rotate Left k cell
def rotate_left(source, k): # Hint, You can write a function for left rotate once and then use it
  for i in range(k):
    temp=source[0]
    for x in range(1,len(source)):
      source[x-1]=source[x]
    source[len(source)-1]=temp
  return source


# Test 03: Shift Right k cell
def shift_right(source, k):   # Hint, You can write a function for right shift once and then use it
  for i in range(len(source)-1,0,-1):
    source[i]=0
    source[i]= source[i-k]
  for x in range(k):
    source[x]=0


  return source



# Test 04: Rotate Right k cell
def rotate_right(source, k):  # Hint, You can write a function for right rotate once and then use it
  for j in range(k):
    temp= source[len(source)-1]
    for i in range(len(source)-1,0,-1):
        source[i]= source[i-1]
        source[i-1]= temp

  return source



# Test 05: Remove an element from an array
def remove(source, idx):
  if idx<=len(source)-1:
    if len(source)-1==idx:
      source[-1]=0
    else:
      for j in range(idx+1,len(source)-1):
          source[j-1]=source[j]
          source[j]=0
    
  return source



# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
  for i in range(len(source)-1,0,-1):
    if source[i]==element:
      for j in range(i,len(source)-1):
        source[j]=source[j+1]

  return source



# Test 07: Splitting an Array
def split_array(a):
  l_sum=0
  r_sum=0
  flag=False
  for i in range(0,len(a)-1):
    l_sum+=a[i]
    for j in range(len(a)-1,i+1,-1):
      r_sum+=a[j]
      if l_sum==r_sum:
        flag = True
            
  return flag


# Test 08: Max Bunch Count
def max_bunch(a):
  # TO DO
  max=0
  counter=1
  for i in range(len(a)-1):
      if a[i]==a[i+1]:
          counter+=1
      else:
          counter=1
      if counter>max:
          max=counter
   
  return max
```
