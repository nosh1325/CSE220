```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    # To Do
    self.head= Node(a[0],None,None)
    h=self.head
    h.next=h
    h.prev=h
    tail=h
    
    for i in range(1,len(a)):
      n= Node(a[i],h,tail)
      tail.next=n
      tail=n

    tail.next=h
    h.prev=tail
      

  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    h=self.head.next
    node_count=1
    while h!=self.head:
      node_count+=1
      h=h.next
    return node_count
    
  
  # prints the elements in the list
  def forwardprint(self):
    n=self.head
    while n!=self.head.prev:
      print(n.element, end= ", ")
      n=n.next
    print(self.head.prev.element)
    

  # prints the elements in the list backward
  def backwardprint(self):
    n=self.head.prev
    while n!=self.head:
      print(n.element, end = ", ")
      n=n.prev
      
    print(self.head.element)
   
    

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    
    if idx<0 or idx>=self.countNode():
      print("Index error")
    
    elif idx==0:
      x=self.head
      return x
    else:
      x=self.head
      x=self.head.next
      idx_counter=1
      while x!=self.head: 
        if idx_counter==idx:
          return x
        x=x.next
        idx_counter+=1

  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    x=self.head
    counter=0
    if elem==x.element:
      return counter
    else:
      while x.next!=self.head:
          if x.element == elem:
              return counter
          x=x.next
          counter+=1
      if elem==x.element:
        return counter
      else:
        return -1 

    

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):

    if idx<0 or idx> self.countNode():
      print("Invalid index")
    else:
      new= Node(elem, None, None)
      if idx==0:
        pred = self.head.prev
        pred.next = new
        new.prev= pred
        new.next = self.head
        self.head.prev = new
        self.head = new
        
      else:
        pred= self.nodeAt(idx-1)
        suc = pred.next
        pred.next=new
        new.next=suc
        new.prev=pred
        suc.prev=new

        
      
    
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    if idx<0 or idx>=self.countNode():
      return None
    else:
      if idx==0:
        curr=self.head
        self.head.prev.next=self.head.next
        self.head.next.prev=self.head.prev
        self.head= curr.next
      else:
        curr= self.nodeAt(idx)
        pred= self.nodeAt(idx-1)
        succ = curr.next
        pred.next=succ
        succ.prev=pred  
    
    return f"{curr.element}"
    
```
